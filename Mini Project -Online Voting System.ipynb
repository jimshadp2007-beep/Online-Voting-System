```python
voters = {
    "jimshad4565": False,
    "shihad3989": False,
    "shameel1801": False,
    "anshad3762": False,
    "murshid7654": False,
    "sree hari6756": False,
    "shameem3524": False,
    "afshan7533": False
}

candidates = {
    "1": {"name": "anishh", "votes": 0},
    "2": {"name": "shamsiya", "votes": 0}
    "3": {"name": "shamna", "votes": 0}
}
1

current_voter = None

# Function for voter login
def voter_login():
    global current_voter6
    voter_id = input("Enter your Voter ID: ")

    if voter_id in voters:
        if voters[voter_id]:
            print("You have already voted. Duplicate voting is not allowed.")
            return False
        else:
            current_voter = voter_id
            print("Login successful.")
            return True
    else:
        print("Invalid Voter ID.")
        return False

# Display candidate list
def display_candidates():
    print("\nCandidate List:")
    for cid, details in candidates.items():
        print(f"{cid}. {details['name']}")

# Cast vote
def cast_vote():
    display_candidates()
    choice = input("Enter candidate number to vote: ")

    if choice in candidates:
        candidates[choice]["votes"] += 1
        voters[current_voter] = True
        print("Vote cast successfully. Thank you for voting!")
    else:
        print("Invalid choice.")

# Display voting results
def display_results():
    print("\nVoting Results:")
    for details in candidates.values():
        print(f"{details['name']}: {details['votes']} votes")

# Vote count summary
def vote_summary():
    total_votes = 0
    for details in candidates.values():
        total_votes += details["votes"]
    print(f"\nTotal Votes Cast: {total_votes}")

# Main menu
def voting_menu():
    while True:
        print("""
------- Online Voting System -------
1. Login
2. Display Candidates
3. Cast Vote
4. Display Results
5. Vote Count Summary
6. Exit
-----------------------------------
""")
        choice = input("Enter your choice (1-6): ")

        if choice == "1":
            voter_login()
        elif choice == "2":
            display_candidates()
        elif choice == "3":
            if current_voter:
                cast_vote()
            else:
                print("Please login first.")
        elif choice == "4":
            display_results()
        elif choice == "5":
            vote_summary()
        elif choice == "6":
            print("Exiting Voting System. Goodbye!")
            break
        else:
            print("Invalid choice. Try again.")

# Program execution
voting_menu()
```
